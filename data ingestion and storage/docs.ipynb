## data type
- structured data
    * data that is organized in a defined maner or schema, typicalle found in relational database
    * easily queryable
    * organized in rows and colums
    * Has a consistent structure
    * example: mysql, postgreSQL, csv, excel,...
- unstructured data
    * data that doesn't have predefinded structure or schema
    * not easily queryable without preprocessing
    * various format
    * example: text file, video and audio, image, email,...
- semi-structured data
    * data that is not as organized as struture data but has some lever of structure in the forme format of tag, hierarchies
    * example: xml, json, logfile
