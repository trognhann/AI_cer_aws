## Sagemaker
1.  Define: fully managed service
    Build, train, deploy model AI
    Flow: collect data & preparation data ==> build and train model ==> deploy ==> monitor
2. build-n algorithms (các thuật toán được tích hợp sẵn)
    Supervised learning: linear regression, clasification 
    Unsupervise learning: PCA (giảm chiều dữ liệu), k-means, Anomaly detection (phát hiện gian lận)
    NLP, CV
3. Automatic model tuning
    tự động thử nghiệm với các hyperparametes khác nhau để tối ưu mô hình
    tối ưu, early stop condition
4. Deployment Option
    real-time inference:payload max 6MB
    serverless inference: dùng cho lưu lượng ko ổn định, lâu ở lần chạy đầu tiên
    Asynchronous inference: near real time large 1GB
    batch transform: High latency, Entire Dataset.

## Federate learning
-  đưa code đến nơi có dữ liệu, chứ ko đưa dữ liệu đến nơi code, nhằm bảo mật dữ liệu
- là kỹ thuật train model phân tán trên cách thiết bị cạnh, (EDGE device)
- flow: gửi model đến các thiết bị tại client ==> client tự huấn luyện dữ liệu của mình ==> gửi trọng số từ client về sever ==> tạo ra model thông minh hơn

    - data private
    - tuân thủ pháp lý GDPR, HIPAA
    - tiết kiệm băng thông
    - low latency
